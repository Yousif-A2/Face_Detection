## **First Step**: Upload two files 
 1.   haarcascade_frontalface_default.xml
 2.   Test image





```python
imagePath = "/content/abba.png"
cascPath = "/content/haarcascade_frontalface_default.xml"
```

## **Second Step**:  Classifier Training  
*   Pass image to the Cascade Classifier Training
*   create the cascade 
*   initialize it with our face cascade
*   Loads the face cascade into memory


```python
import cv2
from google.colab.patches import cv2_imshow
# Create the haar cascade
faceCascade = cv2.CascadeClassifier(cascPath)

# Read the image
gray = cv2.imread(imagePath)
cv2_imshow(gray)
```

## **Third Step**: Face Detection
*   Read the image
*   Convert it to grayscale
*   Many operations in OpenCV are done in grayscale

```python
# Detect faces in the image
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30)
    )

print(faces)
```

## **Fifth Step**: Print the results
*   Print the number of faces in the test image as text 
*   Display test image with faces detection



```python
from google.colab.patches import cv2_imshow
import cv2
# Draw a rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(gray, (x, y), (x+w, y+h), (255, 0,0), 4)
cv2_imshow(gray)
```
